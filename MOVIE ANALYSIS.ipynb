```python
#converting genre list to str
df['Genre'] = df['Genre'].apply(lambda x: x[1:-1])
```

```python
df
```

```python
T = df.Genre.str.split(',',expand=True).iloc[:,0:2]
T.columns=['Genre1','Genre2']
T.Genre2.fillna(T.Genre1,inplace=True)
T
```

```python
#concatenat T dataframe to main dataframe
df = pd.concat([df,T],axis=1)
df
```

```python
df.groupby(['Genre1','Genre2']).World_Sales_In_Dollar.mean().sort_values(ascending=False).head(5)
```

```python
cmap = plt.get_cmap('Pastel1')
colors = [cmap(i) for i in np.linspace(0, 1, 8)]

import matplotlib.pyplot as plt
ax = df.groupby(['Genre1','Genre2']).World_Sales_In_Dollar.mean().sort_values(ascending=False).head(5).plot(kind='bar', figsize=(20,10), color=colors)

plt.title('Bar Chart showing Genre1,Genre2 on basis of world sales')
```

# We can infer that Adventure and Family type of Genre is mostly sale in the world

```python
df.groupby(['Genre1','Genre2']).Domestic_Sales_In_Dollar.mean().sort_values(ascending=False).head(5)
```

```python
cmap = plt.get_cmap('Pastel1')
colors = [cmap(i) for i in np.linspace(0, 1, 8)]

import matplotlib.pyplot as plt
ax = df.groupby(['Genre1','Genre2']).Domestic_Sales_In_Dollar.mean().sort_values(ascending=False).head(5).plot(kind='bar', figsize=(20,10), color=colors)

plt.title('Bar Chart showing Genre1,Genre2 on basis of domestic sales')
```
